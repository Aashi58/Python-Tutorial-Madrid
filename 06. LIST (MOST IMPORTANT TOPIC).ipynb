```python
#find the odd and even
ls=[]
ls_e=[]
ls_o=[]
for i in range(10):
  n=int(input("enter the number"))
  ls.append(n)
for i in ls:
  if i%2==0:
    ls_e.append(i)
  else:
    ls_o.append(i)
print(ls)
print(ls_e)
print(ls_o)
```

```python
#find the minimum number
ls=[]
for i in range(5):
  n=int(input())
  ls.append(n)
print(ls)
m=ls[0]
for i in ls:
  if i<m:
    m=i
print("minimum is",m)
```

```python
#find the smallest, greater and equal number
ls=[]
cg=0
cs=0
ce=0
for i in range(10):
  n=int(input())
  ls.append(n)
print(ls)
choice=int(input("enter the number for comparison"))
for i in ls:
  if i>choice:
    cg=cg+1
  elif i>choice:
    cs=cs+1
  else:
      ce=ce+1
print("count greater",cg)
print("count small",cs)
print("count equal",ce)
```

```python
#find the occerence of the element in the list with position
ls=[]
for i in range(10):
  n=int(input())
  ls.append(n)
print(ls)
co=0
n1=int(input("enter the number to be search"))
for i in range(len(ls)):
  if n1==ls[i]:
    co=co+1
    if co>0:
      print(n1,"is found at",i+1,"position")
```

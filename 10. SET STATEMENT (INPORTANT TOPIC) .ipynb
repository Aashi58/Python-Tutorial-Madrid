```python
# set()
s=set('hello')
print(s)
```

```python
s=set((1,2,3,4,5))   # convert tuple to set
```

```python
s=set()    # create an empty set
s.add(10)
s.add(20)
s.add(30)
print(s)
primeNums={2,3,5,7}
s.update(primeNums)  # update set with another set
print(s)
s.remove(2)    #remove an element
print(s)
```

```python
# SET CLEAR
langs={'python','c++','c','go'}
langs.clear()
print('set after the clear() method is called:',langs)
```

```python
# SET COPY
langs={'python','c++','c','go'}
copiedlangs=langs.copy()
print("original set:",langs)
print("copied set:",copiedlangs)
```

```python
#SET DIFFERENCE
nums1={1,2,3,4,5,2}
nums2={4,5,6,7,8,9}
nums3=nums1.difference(nums2)
nums4=nums2.difference(nums1)
print("nums1-nums2:",nums3)
print("nums2-nums1:",nums4)
```

```python
# SET INTERSATION
nums={1,2,3,4,5}
oddNums={1,3,5,7,9}
commonNums=nums.intersection(oddNums)
print("common element in nums and oddNumns are:",commonNums)
```

```python
#
nums={1,2,3,4,5}
oddNums={1,3,5,7,9}
primeNums={2,3,5,7}
commonNums=nums.intersection(oddNums, primeNums)
print("common elements:",commonNums)
```

```python
# SET UNION
nums1={1,2,2,3,4,5}
nums2={4,5,6,7,7,8}
distinct_nums=nums1.union(nums2)
print("the union of two sets is:",distinct_nums)
```

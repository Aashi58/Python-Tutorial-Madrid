```python
#pop method
romanNums={'I':1,'II':2,'III':3}
deletedvalue=romanNums.pop('III')
print('the popped element is: ', deletedvalue)
print('updated dictionary: ',romanNums)
```

```python
#delete method
romanNums={'I':1,'II':2,'III':3}
deletedvalue=romanNums.pop('III')
print('the popped element is: ', deletedvalue)
```

```python
# not found use when we dont have that value
romanNums={'I':1,'II':2,'III':3}
deletedvalue=romanNums.pop('IV','not found')
print('the popped element is: ', deletedvalue)
print('updated dictionary: ',romanNums)
```

```python
#update method
romanNums={'I':1,'II':2,'III':3,}
print("dictioanry:", romanNums)
evenRomanNums={'IV':4,'V':5,'VI':6}
romanNums.update(evenRomanNums)
print("update Dictionary:",romanNums)
```

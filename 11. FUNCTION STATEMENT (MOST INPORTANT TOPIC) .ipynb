```python
# RECURSION
#SUM OF N NATURAL NUMBER
def sum(n):
  if n==1:
    return 1
  else:
    return n+sum(n-1)
print(sum(7))
```

```python
#factorial
def fact(n):
  if n==0:
    return 1
  elif n==1:
    return 1
  else:
    return n*fact(n-1)
print(fact(7))
```

```python
#fabonacci
f=0
s=1
for i in range(10):
  if i<=1:
    next=i
  else:
    next=f+s
    f=s
    s=next
  print(next,end=" ")
```

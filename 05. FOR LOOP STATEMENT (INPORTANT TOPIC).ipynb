```python
#prime number

for i in range(1,100):
  n=i
  j=1
  co=0
  while j<=n:
    if n%j==0:
      co=co+1
    j=j+1
  if co==2:
    print(i,end=" ")
```

```python
#palindrome number in a given range

for i in range(11,999):
  n=i
  rev=0
  while n>0:
    rev=rev*10+n%10
    n=n//10
  if rev==i:
    print(i,end=" ")
```

```python
#armstrong

for i in range(100,999):
  n=i
  sum=0
  while n>0:
    rem=n%10
    sum=sum+rem**3
    n=n//10
  if sum==i:
    print(i,end=" ")
```
